```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
from google3.colab import files
uploaded = files.upload()
```

```python
df=pd.read_csv('temperatures.csv');
```

```python
df
```

```python
x=df['YEAR']
```

```python
y=df['ANNUAL']
y=df['JAN'] #particular month
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,test_size=0.25,random_state=2);
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train.values.reshape(-1,1),y_train);
```

```python
y_predict=lr.predict(x_test.values.reshape(-1,1))
```

```python
y_predict
```

```python
mse =mean_absolute_error(y_test,y_predict)
```

```python
mse
```

```python
mae=mean_absolute_error(y_test,y_predict)
```

```python
mae
```

```python
r2_score=r2_score(y_test,y_predict)
```

```python
r2_score
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_predict)
plt.xlabel("Temperature")
plt.ylabel("Year")
plt.show()
```
